```python
def selectionsort(arr):
    arr=[18,15,17,16,24,14,19]

    for i in range(len(arr)):
        min_index=i
        for j in range(i+1,len(arr)):
            if arr[j]<arr[min_index]:
                min_index=j
        arr[i],arr[min_index]=arr[min_index],arr[i]
    return arr

print(selectionsort(arr))
```

```python
arr=[6,8,10,2,5]
def insertion(arr):
    n=len(arr)
    for i in range(1,n):
        key=arr[i]
        j=i-1
        while j>=0 and arr[j]>key:
            arr[j+1]=arr[j]
            j=j-1
        arr[j+1]=key
insertion(arr)
print(arr)
```
